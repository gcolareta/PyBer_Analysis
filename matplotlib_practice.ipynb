```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# create the scatter plot
plt.plot(y_axis, x_axis, 'o', c="red", label="Chicago")

# Annotate the plot
plt.gca().invert_yaxis()
plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.title("Pyber Fare by Month")
plt.xlim(0,50)
plt.legend()
```

```python
# Bubble chart

plt.scatter(x_axis, y_axis, s=y_axis)

# Make bubbles bigger but stil in relation to y_axis
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

# use the y_axis_larger variable for the size
plt.scatter(x_axis, y_axis, s=y_axis_larger)




```

```python
# Make a bubble plot using "list comprehension" to make bubbles bigger depending on y_axix value
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Bubble chart skill drill

fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], c="skyblue", alpha=0.2, edgecolor="black", linewidth=2, label="Boston")

# Annotate the plot
plt.gca().invert_yaxis()
plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.title("Pyber Fare by Month")
plt.xlim(0,50)
plt.legend()
```

```python
# Pie Charts

# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()

```

```python
# Object-oriented method

fig, ax = plt.subplots(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "orangered", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,labels=x_axis, explode=explode_values, colors=colors, autopct='%.1f%%', shadow=True, startangle=90, counterclock=False )
plt.show()
```
